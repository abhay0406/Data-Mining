# 1.4 Apply K-Means clustering on scaled data and determine optimum clusters. Apply elbow curve and silhouette score. Explain the results properly. Interpret and write inferences on the finalized clusters.

```python
from sklearn.cluster import KMeans
```

```python
scaled_bank.describe()
```

```python
# determining the number of clusters using inertia function in kmeans
```

```python
k_means = KMeans(n_clusters = 1)
k_means.fit(scaled_bank)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 2)
k_means.fit(scaled_bank)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 3)
k_means.fit(scaled_bank)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 4)
k_means.fit(scaled_bank)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 5)
k_means.fit(scaled_bank)
k_means.inertia_
```

```python
# using inertia fucntion we conclude that 3 would be the ideal number of clusters because the drop in inertia value after 
# 3 number of clusters are not significant.
```

```python
#getting inertia value using WSS method
wss =[] 
for i in range(1,11):
    KM = KMeans(n_clusters=i,random_state=1)
    KM.fit(scaled_bank)
    wss.append(KM.inertia_)
wss
```

```python
# plotting WSS valuesti determine the ideal number of clusters
plt.plot(range(1,11), wss)
```

```python
# we have determined that 3 is the idea number of clusters
# getting labels of each observation for 3 clusters
k_means = KMeans(n_clusters = 3)
k_means.fit(scaled_bank)
labels = k_means.labels_
labels
```

```python
# adding cluster labels to original data
bank["Clus_kmeans1"] = labels
bank.head(5)
```

```python
# checking silhoutte score
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
# adding sil width column to original data set
bank["sil_width"] = sil_width
bank.head(5)
```
